````python
betteredgeWords = nltk.word_tokenize(betteredge)
````

````python
len(betteredgeWords)
````

## Concordances

A concordance is sometimes called "keywords-in-context" 

````python
betteredgeText.concordance('diamond', lines=154)
````

````python
betteredgeText.collocation_list()
````

````python
betteredgeText.collocation_list(num=20, window_size=3)
````

````python
betteredgeText.dispersion_plot(['Crusoe*', 'pipe'])
````

````python
betteredgeText.findall('<diamond>')
````

````python
betteredgeText.count('Diamond')
````

````python
betteredgeText.generate()
````

````python
betteredgeText.readability('kinkaid')
````

````python
betteredgeText.similar('diamond')
````

````python
betteredgeText.vocab()
````

````python
betteredgeVocab = betteredgeText.vocab()
````

````python
help(betteredgeVocab)
````

````python
betteredgeVocab.freq('diamond')
````

````python
betteredgeVocab.freq('Diamond')
````

````python
betteredgeVocab.hapaxes()
````

````python
betteredgeVocab.plot(40)
````
