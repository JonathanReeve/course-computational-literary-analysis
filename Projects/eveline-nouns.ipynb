```python
import spacy
from collections import Counter
from nltk.text import Text
from nltk import word_tokenize
```

```python
nlp = spacy.load('en_core_web_lg')
```

```python
eveline = nlp(open('eveline.md').read())
```

```python
evelineNouns = [word.string.lower().strip()
               for word in eveline
               if word.pos_ == 'NOUN']
```

```python
Counter(list([w.string.lower().strip() for w in 
              eveline.noun_chunks])).most_common()
```

```python
Counter(evelineNouns).most_common()
```

```python
evelineTokens = word_tokenize(open('eveline.md').read())
```

```python
evelineText = Text(evelineTokens)
```

```python
evelineText.concordance('father', width=70)
```

```python
evelineText.concordance('mother', width=70)
```
